```python
from googletrans import Translator, LANGUAGES

def translate_to_multiple_languages(sentence, target_languages):
    translator = Translator()
    translations = {}

    for lang in target_languages:
        lang = lang.strip()  # Remove any leading/trailing whitespace
        if lang in LANGUAGES.keys():  # Check if the language is supported
            try:
                output = translator.translate(sentence, dest=lang)
                translations[lang] = output.text
            except Exception as e:
                translations[lang] = f"Error: {str(e)}"
        else:
            translations[lang] = "Error: Language not supported"
    return translations

sentence = input("Enter the sentence: ")
target_languages = input("Enter target languages separated by commas (e.g., en, fr, es): ").split(',')

translations = translate_to_multiple_languages(sentence, target_languages)

# Display the translations or errors
for lang, translation in translations.items():
    print(f"Translation to {lang}: {translation}")

# Show a list of supported languages
print("\nSupported languages:")
for lang_code, lang_name in LANGUAGES.items():
    print(f"{lang_code}: {lang_name.capitalize()}")
```
